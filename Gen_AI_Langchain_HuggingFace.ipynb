## Through API End Points

```python
from langchain_huggingface import HuggingFaceEndpoint

llm = HuggingFaceEndpoint(
    repo_id="mistralai/Mistral-7B-Instruct-v0.3",
    task="text-generation",
    max_new_tokens=100,
    do_sample=False,
)
```

```python
llm.invoke("What is Machine learning")
```

**Purpose of LLMChain**

The LLMChain helps combine a prompt template (or any other logic that involves prompt creation) with a language model to create a chain of actions. It allows you to structure interactions with the language model in a way that can be reused and composed easily in complex applications.

It can be thought of as a high-level interface to manage the input to and output from the language model.

```python
from langchain import PromptTemplate, LLMChain

question="Prime minister of India in 2022"
template = """Question: {question}
Answer: Let's think step by step."""
prompt = PromptTemplate(template=template, input_variables=["question"])
print(prompt)
```

```python
llm_chain=LLMChain(llm=llm,prompt=prompt)
print(llm_chain.invoke(question))
```

## Hugging Face Pipeline

* For hugging face pipeline you have to install the model locally.
* It makes the process faster, maximum privacy, speed, and flexibility, can use without Internet.
* Don't Need API To call as it is installed.



Among transformers, the Pipeline is the most versatile tool in the Hugging Face toolbox. LangChain being designed primarily to address RAG and Agent use cases, the scope of the pipeline here is reduced to the following text-centric tasks: “text-generation", “text2text-generation", “summarization”, “translation”.
Models can be loaded directly with the **from_model_id** method


```python
# Import HuggingFacePipeline from langchain_huggingface
# This wraps a Hugging Face pipeline to be used easily in LangChain workflows.
from langchain_huggingface import HuggingFacePipeline

# Import AutoModelForCausalLM from transformers
# This is used to automatically load a model for **causal language modeling** (text generation models like GPT-2, LLaMA, etc.)
from transformers import AutoModelForCausalLM

# Import AutoTokenizer from transformers
# This is used to load the **tokenizer** for the model.
# Tokenizer converts input text into tokens (numbers) that models can understand, and vice versa.
from transformers import AutoTokenizer

# Import the generic pipeline utility from transformers
# This creates a ready-to-use **pipeline** object for tasks like text-generation, summarization, translation, etc.
from transformers import pipeline

```

```python
# "gpt2" is the name of a small, pre-trained language model available on Hugging Face.
model_id = "gpt2"

# This will download the model weights and configuration needed for text generation tasks.
model = AutoModelForCausalLM.from_pretrained(model_id)

# Load the corresponding tokenizer for the model
# The tokenizer is necessary to encode input text into tokens and decode output tokens back to text.
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
!pip install huggingface_hub[hf_xet]
```

```python
# Create a text-generation pipeline using the loaded model and tokenizer
# The pipeline will handle encoding inputs, running the model, and decoding outputs automatically
# 'max_new_tokens=100' means it will generate up to 100 new tokens in the output text
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=100
)

# Wrap the Hugging Face pipeline using LangChain's HuggingFacePipeline
# This allows the pipeline to be used as an LLM (Large Language Model) inside LangChain applications
hf = HuggingFacePipeline(pipeline=pipe)
```

```python
hf.invoke("What is machine learning..")
```

```python
## Use HuggingfacePipelines With Gpu
gpu_llm = HuggingFacePipeline.from_model_id(
    model_id="gpt2",
    task="text-generation",
    device=0,  # replace with device_map="auto" to use the accelerate library.
    pipeline_kwargs={"max_new_tokens": 100},
)
```

```python
from langchain_core.prompts import PromptTemplate

template = """Question: {question}

Answer: Let's think step by step."""
prompt = PromptTemplate.from_template(template)
```

```python
chain=prompt|gpu_llm
question="What is artificial intelligence?"
chain.invoke({"question":question})
```
